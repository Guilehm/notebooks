```python
import random
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('source/airplanes_ocurrences/oco.csv')
```

# Head of data frame

```python
df.head()
```

# Convert from int64 to str for slicing

```python
df['codigo_ocorrencia'] = df['codigo_ocorrencia'].apply(str)
```

# Split and store date values converting to integers

```python
df['ano_ocorrencia'] = df['codigo_ocorrencia'].str[:4].astype(int)
df['mes_ocorrencia'] = df['codigo_ocorrencia'].str[4:6].astype(int)
df['dia_ocorrencia'] = df['codigo_ocorrencia'].str[6:8].astype(int)
```

#  Find occurences in 2018

```python
df.loc[df['ano_ocorrencia'] == 2018]
```

# Count ocurrences by year

```python
ocurrences_by_year = df.ano_ocorrencia.value_counts()
ocurrences_by_year
```

# Reset index and change column names

```python
ocurrences_by_year = ocurrences_by_year.reset_index()
ocurrences_by_year.columns = ('ano', 'contagem')
ocurrences_by_year
```

# Generate graph 

```python
plt.figure(figsize=(15, 8))
plt.bar(ocurrences_by_year.ano, ocurrences_by_year.contagem)
plt.xlabel('ano', fontsize=30)
plt.ylabel('quantidade', fontsize=30)
plt.xticks(ocurrences_by_year.ano)
plt.tick_params(axis='both', labelsize=13)
plt.title('Acidentes x Ano', fontsize=40)
```
